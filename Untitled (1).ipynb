```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.facecolor'] = 'pink'
```

```python
data = pd.read_csv('world happiness report 2019.csv')
```

```python
data.head()
```

```python
data_columns = ['Country_or_region','Score','GDP','Social_support','Healthy_life_expectancy','Freedom_to_make_life_choices','Generosity','Perceptions_of_corruption']
```

```python
data = data[data_columns].copy()
```

```python
happy_df = data.rename({'Score':'Score_'})
```

```python
happy_df.isnull().sum()
```

```python
#plot between happiness score and GDP per capita

plt.rcParams['figure.figsize'] = (15,8)
plt.title('Plot between Happiness Score and GDP')
sns.scatterplot(x = happy_df.Score,y = happy_df.GDP, hue = happy_df.Country_or_region, s = 200);

plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP')
plt.show()

```

```python
#total countries

total_country = happy_df.groupby('Country_or_region')[['GDP']].count()
print(total_country)
```
